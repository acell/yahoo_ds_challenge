```python
import sklearn
import csv
import math
import numpy as np
import pandas as pd
import scipy.io.wavfile
import os
import random
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
# maybe run other libraries just to get more features for prediction, if that makes sense
# not like actually using those other libraries as my model

# ok, hot

# get github working
# prediction accuracy
# do validation to say what we predict accuracy to be
# data cleaning, feature engineering 
```

```python
train_files = []
for filename in os.listdir(str(os.getcwd())+'/umich_ds_cc_2017-master/train_data'):
    train_files.append(filename)
```

```python
# load the training files
file_dict = {}
for file in train_files:
    file_dict[file] = {}
    file_dict[file]['length'] = scipy.io.wavfile.read('umich_ds_cc_2017-master/train_data/'+str(file))[0]
    file_dict[file]['contents'] = scipy.io.wavfile.read('umich_ds_cc_2017-master/train_data/'+str(file))[1]
```

```python
file_dict
```

```python
# clean the data + build in additional features
wav_dict = pd.DataFrame.from_dict(file_dict, 'index')
# wav_dict['mean'] = pd.Series((wav_dict['contents'][x].mean() for x in np.arange(len(wav_dict))), index=wav_dict.index)
# wav_dict['median'] = pd.Series((np.median(wav_dict['contents'][x]) for x in np.arange(len(wav_dict))), index=wav_dict.index)
# wav_dict['std'] = pd.Series((wav_dict['contents'][x].std() for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['min'] = pd.Series((wav_dict['contents'][x].min() for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['max'] = pd.Series((wav_dict['contents'][x].max() for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['range'] = wav_dict['max'] - wav_dict['min']
wav_dict['abs_range'] = abs(abs(wav_dict['max']) - abs(wav_dict['min']))
wav_dict['abs_mean'] = pd.Series((abs(wav_dict['contents'][x]).mean() for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['abs_median'] = pd.Series((np.median(abs(wav_dict['contents'][x])) for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['abs_std'] = pd.Series((abs(wav_dict['contents'][x]).std() for x in np.arange(len(wav_dict))), index=wav_dict.index)
wav_dict['abs_range_pct'] = wav_dict['abs_range'] / wav_dict['max'] * 100
wav_dict
```

```python
# trying to improve my model so it doesn't think everything is a fuckin marimba
for index, row in wav_dict.iterrows():
    same = 0
    dec = 0
    inc = 0
    
    dec_streak = 0 
    inc_streak = 0
    same_streak = 0
    max_dec_streak = 0
    max_inc_streak = 0
    max_same_streak = 0
    
    maj_inc = 0
    maj_dec = 0
    kinda_maj_inc = 0
    kinda_maj_dec = 0
    very_maj_inc = 0
    very_maj_dec = 0
    for ind in np.arange(3000):
        if (row['contents'][ind] > row['contents'][ind - 1]):
            inc += 1
            inc_streak += 1
            dec_streak = 0
            same_streak = 0
            if (inc_streak > max_inc_streak):
                max_inc_streak = inc_streak
                
        elif (row['contents'][ind] < row['contents'][ind - 1]):
            dec += 1
            inc_streak = 0
            dec_streak += 1
            same_streak = 0
            if (dec_streak > max_dec_streak):
                max_dec_streak = dec_streak
        else:
            same += 1
            inc_streak = 0
            dec_streak = 0
            same_streak += 1
            if (same_streak > max_same_streak):
                max_same_streak = same_streak
                
        if (row['contents'][ind] > (row['contents'][ind - 1] + 100)):
            kinda_maj_inc += 1
        elif (row['contents'][ind] < (row['contents'][ind - 1] - 100)):
            kinda_maj_dec += 1
            
        if (row['contents'][ind] > (row['contents'][ind - 1] + 500)):
            maj_inc += 1
        elif (row['contents'][ind] < (row['contents'][ind - 1] - 500)):
            maj_dec += 1
            
        if (row['contents'][ind] > (row['contents'][ind - 1] + 1000)):
            very_maj_inc += 1
        elif (row['contents'][ind] < (row['contents'][ind - 1] - 1000)):
            very_maj_dec += 1
            
    wav_dict.loc[index, 'same'] = same
    wav_dict.loc[index, 'inc'] = inc
    wav_dict.loc[index, 'dec'] = dec
    
    wav_dict.loc[index, 'maj_inc'] = maj_inc
    wav_dict.loc[index, 'maj_dec'] = maj_dec
    wav_dict.loc[index, 'kinda_maj_inc'] = kinda_maj_inc
    wav_dict.loc[index, 'kinda_maj_dec'] = kinda_maj_dec
    wav_dict.loc[index, 'very_maj_inc'] = very_maj_inc
    wav_dict.loc[index, 'very_maj_dec'] = very_maj_dec
    
    wav_dict.loc[index, 'same_streak'] = max_same_streak
    wav_dict.loc[index, 'inc_streak'] = max_inc_streak
    wav_dict.loc[index, 'dec_streak'] = max_dec_streak
```

```python
y = pd.Series([wav_dict.index[x].split('_')[0] for x in np.arange(len(wav_dict))], index=wav_dict.index)
X = wav_dict
```

```python
accuracy = 0
for x in np.arange(30):
    X_train, X_test, y_train, y_test = train_test_split(
                                       X, y, test_size=0.2, random_state=random.randint(0, 100), stratify=y)
    xtra = ExtraTreesClassifier(n_estimators=50)
    xtra.fit([list(X_train.drop('contents', axis=1).iloc[x]) for x in np.arange(len(X_train))], np.array(y_train))
    xtras = xtra.predict([list(X_test.drop('contents', axis=1).iloc[x]) for x in np.arange(len(X_test))])
    accuracy += (np.array(y_test) == xtras).mean()
    print((np.array(y_test) == xtras).mean())
print('avg accuracy: ' + str(accuracy / 30))
```
